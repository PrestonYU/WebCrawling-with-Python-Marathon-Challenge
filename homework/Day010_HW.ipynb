## Grab + PyQuery

```python
import logging
logging.basicConfig(level=logging.DEBUG)
from grab import Grab
from pyquery import PyQuery as pq
```

```python
g = Grab()
resp = g.go('https://www.dcard.tw/f')
doc = pq(resp.body)
title = doc('title')
print(type(title), title.text())
```
